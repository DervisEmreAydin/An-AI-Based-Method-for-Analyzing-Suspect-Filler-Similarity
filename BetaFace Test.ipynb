```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
model_name = "VGG-Face"
distance_metric = "cosine"
```

```python
suspect_img = "Target\\DervisEmreAydin.jpg"
betaface_img1 = "BetaFace\\1.jpg"

Suspect_img = cv2.imread(suspect_img)
Betaface_img1 = cv2.imread(betaface_img1)

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(Suspect_img[:,:,::-1])
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(Betaface_img1[:,:,::-1])

verify1= DeepFace.verify(suspect_img, betaface_img1,
                         model_name= "VGG-Face",
                         distance_metric= "cosine")

verify1_list = [verify1.get('distance'),
                verify1.get('model'),
                verify1.get('similarity_metric')]
print(verify1_list)
```
